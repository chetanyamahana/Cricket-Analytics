```python
import numpy as np
import pandas as pd
```

```python
batsman_df=pd.read_excel('6655876402143_batsman_round1 (1).xlsx')
bowler_df=pd.read_excel('6655876ef0ec2_bowler_round1.xlsx')
matches_df=pd.read_excel('6655877834f04_matchlevel_round1.xlsx')
```

```python
# Function to remove duplicates
def remove_duplicates(df):
    return df.drop_duplicates(keep='last')

# Apply the function to each dataframe
matches_df = remove_duplicates(matches_df)
batsman_df = remove_duplicates(batsman_df)
bowler_df = remove_duplicates(bowler_df)

# Check if duplicates are removed
print(matches_df.duplicated().sum())
print(batsman_df.duplicated().sum())
print(bowler_df.duplicated().sum())
```

```python
matches_df = matches_df[(matches_df['match id'] >= 8000000) & (matches_df['match id'] <= 10000000)]
batsman_df = batsman_df[(batsman_df['match id'] >= 8000000) & (batsman_df['match id'] <= 10000000)]
bowler_df = bowler_df[(bowler_df['match id'] >= 8000000) & (bowler_df['match id'] <= 10000000)]
```

```python
batsman_df = batsman_df[(batsman_df['batsman_id'] >= 36000) & (batsman_df['batsman_id'] <= 10000000)]
```

```python
for index, row in batsman_df.iterrows():
    if pd.isnull(row['strike_rate']) and not pd.isnull(row['runs']) and not pd.isnull(row['balls_faced']) and row['balls_faced'] != 0:
        batsman_df.at[index, 'strike_rate'] = (row['runs'] / row['balls_faced']) * 100
```

```python
batsman_df.head()
```

```python
batsman_df.isnull().sum()
```

```python
batsman_df['Fours'].fillna(0, inplace=True)
batsman_df['Sixes'].fillna(0, inplace=True)
```

```python
print(batsman_df.isnull().sum())
```

```python
merged_df = pd.merge(
    batsman_df, bowler_df,
    on=['match id', 'bowler_id'],
    how='left',
    suffixes=('', '_y')
)
```

```python
merged_df
```

```python
columns_to_impute = [
    'wicket kind', 'out_by_bowler', 'out_by_fielder', 
    'bowler_id', 'bowler_dob', 'bowler_details', 
    'is_bowler_keeper', 'is_bowler_captain'
]

for column in columns_to_impute:
    if column in merged_df.columns:
        batsman_df[column].fillna(merged_df[f'{column}'], inplace=True)
```

```python
batsman_df.isnull().sum()
```

```python
null_run_records = batsman_df[batsman_df['runs'].isnull()]

# Display the records
print(null_run_records)
```

```python
mask = batsman_df['runs'].isnull() & batsman_df['strike_rate'].notnull() & batsman_df['balls_faced'].notnull()

# Calculate runs
batsman_df.loc[mask, 'runs'] = (batsman_df.loc[mask, 'strike_rate'] / 100) * batsman_df.loc[mask, 'balls_faced']

# Optionally round the runs to the nearest integer if required
batsman_df['runs'] = batsman_df['runs'].round().astype('Int64')
```

```python
batsman_df.isnull().sum()
```

```python
null_strike_rate_records = batsman_df[batsman_df['strike_rate'].isnull()]

# Display the records
print(null_strike_rate_records)
```

```python
mask_balls_faced_0 = (batsman_df['balls_faced'] == 0) & (batsman_df['strike_rate'].isna())

# Set strike_rate to 0 where balls_faced is 0 and strike_rate is NaN
batsman_df.loc[mask_balls_faced_0, 'strike_rate'] = 0
```

```python
batsman_df['wicket kind'].fillna('Unknown', inplace=True)
```

```python
batsman_df['out_by_bowler'] = batsman_df['out_by_bowler'].fillna('Unknown')
```

```python
# Columns to remove
cols_to_remove = ['match id', 'batsman_id', 'is_batsman_captain', 'is_batsman_keeper', 'inning', 'is_bowler_keeper', 'is_bowler_captain', 'bowler_id']

numerical_cols = list(numerical_cols)

# Remove columns from the list
for col in cols_to_remove:
    if col in numerical_cols:
        numerical_cols.remove(col)
```

```python
numerical_cols
```

```python
for col in numerical_cols:
    Q1 = batsman_df[col].quantile(0.25)
    Q3 = batsman_df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = batsman_df[(batsman_df[col] < lower_bound) | (batsman_df[col] > upper_bound)][col]
    print(f"Outliers in column '{col}': {outliers.values}")
```

```python
for col in numerical_cols:
    Q1 = batsman_df[col].quantile(0.25)
    Q3 = batsman_df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    non_outliers = batsman_df[(batsman_df[col] >= lower_bound) & (batsman_df[col] <= upper_bound)][col]
    mean_value = int(np.round(non_outliers.mean()))  # Round off the mean value
    batsman_df.loc[batsman_df[col] < lower_bound, col] = mean_value
    batsman_df.loc[batsman_df[col] > upper_bound, col] = mean_value
```

```python
# Convert columns to the appropriate datatypes
batsman_df['match id'] = batsman_df['match id'].astype(int)
batsman_df['batsman'] = batsman_df['batsman'].astype('string')
batsman_df['batsman_id'] = batsman_df['batsman_id'].astype(int)
batsman_df['batsman_details'] = batsman_df['batsman_details'].astype('string')
batsman_df['inning'] = batsman_df['inning'].astype(int)
batsman_df['runs'] = batsman_df['runs'].astype('Int64')
batsman_df['balls_faced'] = batsman_df['balls_faced'].astype(int)
batsman_df['over_faced_first'] = batsman_df['over_faced_first'].astype(float)
batsman_df['wicket kind'] = batsman_df['wicket kind'].astype('string')
batsman_df['out_by_bowler'] = batsman_df['out_by_bowler'].astype('string')
batsman_df['out_by_fielder'] = batsman_df['out_by_fielder'].astype('string')
batsman_df['bowler_id'] = pd.to_numeric(batsman_df['bowler_id'], errors='coerce').astype('Int64')  # allows for NULL values
batsman_df['bowler_details'] = batsman_df['bowler_details'].astype('string')
batsman_df['is_bowler_keeper'] = batsman_df['is_bowler_keeper'].astype('Int64').astype(bool, errors='ignore')  # allows for blank values
batsman_df['strike_rate'] = batsman_df['strike_rate'].astype(float)

# Convert Fours and Sixes columns to integers, handling non-integer float values
batsman_df['Fours'] = pd.to_numeric(batsman_df['Fours'], errors='coerce').fillna(0).astype(int).astype('Int64')  # handles NULL and non-integer float values
batsman_df['Sixes'] = pd.to_numeric(batsman_df['Sixes'], errors='coerce').fillna(0).astype(int).astype('Int64')  # handles NULL and non-integer float values
```

```python
# Function to parse dates
def parse_dates(date_str):
    for fmt in ('%Y-%m-%d %H:%M:%S', '%Y-%m-%d'):
        try:
            return pd.to_datetime(date_str, format=fmt)
        except ValueError:
            pass
    return pd.NaT

# Apply the date parsing function
batsman_df['batsman_dob'] = batsman_df['batsman_dob'].apply(parse_dates)
batsman_df['bowler_dob'] = batsman_df['bowler_dob'].apply(parse_dates)
batsman_df['match_dt'] = batsman_df['match_dt'].apply(parse_dates)

# Format the dates to MM/DD/YYYY
batsman_df['batsman_dob'] = batsman_df['batsman_dob'].dt.strftime('%m/%d/%Y')
batsman_df['bowler_dob'] = batsman_df['bowler_dob'].dt.strftime('%m/%d/%Y')
batsman_df['match_dt'] = batsman_df['match_dt'].dt.strftime('%m/%d/%Y')

print(batsman_df[['batsman_dob', 'bowler_dob','match_dt']])
```
