```python
import numpy as np
import pandas as pd
```

```python
bowler_df=pd.read_excel('bowler_round1.xlsx')
```

```python
bowler_df.head()
```

```python
# Function to remove duplicates
def remove_duplicates(df):
    return df.drop_duplicates(keep='last')

# Apply the function to each dataframe
bowler_df = remove_duplicates(bowler_df)

# Check if duplicates are removed
print(bowler_df.duplicated().sum())
```

```python
bowler_df.isnull().sum()
```

```python
def calculate_economy(row):
    if pd.isna(row['economy']) and row['balls_bowled'] != 0 and not pd.isna(row['balls_bowled']):
        return (row['runs'] * 6) / row['balls_bowled']
    return row['economy']

# Function to calculate runs
def calculate_runs(row):
    if pd.isna(row['runs']) and row['balls_bowled'] != 0 and not pd.isna(row['balls_bowled']):
        return (row['economy'] * row['balls_bowled']) / 6
    return row['runs']

# Apply the function to calculate economy rate where the value is null
bowler_df['economy'] = bowler_df.apply(calculate_economy, axis=1)

# Apply the function to calculate runs where the value is null
bowler_df['runs'] = bowler_df.apply(calculate_runs, axis=1)
```

```python
bowler_df.isnull().sum()
```

```python
bowler_df['dots'].fillna(0, inplace=True)
```

```python
bowler_df['wides'].fillna(0, inplace=True)
```

```python
column_names = bowler_df.columns

# Get the data types of all columns
column_dtypes = bowler_df.dtypes

# Combine column names and data types into a DataFrame
column_info = pd.DataFrame({'Data Type': column_dtypes})

print(column_info)
```

```python
bowler_df['match id'] = bowler_df['match id'].astype('int64')
bowler_df['bowler'] = bowler_df['bowler'].astype('string')
bowler_df['bowler_id'] = bowler_df['bowler_id'].astype('int64')
bowler_df['bowler_details'] = bowler_df['bowler_details'].astype('string')
bowler_df['inning'] = bowler_df['inning'].astype('int64')
bowler_df['runs'] = bowler_df['runs'].astype('int64')
bowler_df['wicket_count'] = bowler_df['wicket_count'].astype('int64')
bowler_df['balls_bowled'] = bowler_df['balls_bowled'].astype('int64')
bowler_df['economy'] = bowler_df['economy'].astype('float64')
bowler_df['maiden'] = bowler_df['maiden'].astype('int64')
bowler_df['dots'] = bowler_df['dots'].astype('int64')
bowler_df['Fours'] = bowler_df['Fours'].astype('int64')
bowler_df['Sixes'] = bowler_df['Sixes'].astype('int64')
bowler_df['wides'] = bowler_df['wides'].astype('int64')
bowler_df['noballs'] = bowler_df['noballs'].astype('int64')
```

```python
bowler_df = bowler_df[(bowler_df['match id'] >= 8000000) & (bowler_df['match id'] <= 10000000)]
bowler_df = bowler_df[(bowler_df['bowler_id'] >= 36000) & (bowler_df['bowler_id'] <= 10000000)]
bowler_df = bowler_df[(bowler_df['is_bowler_captain'] == 0) | (bowler_df['is_bowler_captain'] == 1)]
bowler_df = bowler_df[(bowler_df['is_bowler_keeper'] == 0) | (bowler_df['is_bowler_keeper'] == 1)]
bowler_df = bowler_df[(bowler_df['inning'] == 1) | (bowler_df['inning'] == 2)]
bowler_df = bowler_df[(bowler_df['runs'] >= 0) & (bowler_df['runs'] <= 200)]
bowler_df = bowler_df[(bowler_df['wicket_count'] >= 0) & (bowler_df['wicket_count'] <= 10)]
bowler_df = bowler_df[(bowler_df['balls_bowled'] >= 0) & (bowler_df['balls_bowled'] <= 30)]
bowler_df = bowler_df[(bowler_df['economy'] >= 0) & (bowler_df['economy'] <= 36)]
bowler_df = bowler_df[(bowler_df['maiden'] >= 0) & (bowler_df['maiden'] <= 5)]
bowler_df = bowler_df[(bowler_df['dots'] >= 0) & (bowler_df['dots'] <= 22)]

```

```python
bowler_df.head()
```

```python
bowler_df['bowler_dob'] = pd.to_datetime(bowler_df['bowler_dob'], errors='coerce').dt.strftime('%m/%d/%Y')
bowler_df['match_dt'] = pd.to_datetime(bowler_df['match_dt'], errors='coerce').dt.strftime('%m/%d/%Y')
```

```python

# Replace NaT values with '00-00-0000'
bowler_df['bowler_dob'] = bowler_df['bowler_dob'].fillna('00-00-0000')
bowler_df.loc[bowler_df['bowler_dob'] != '00-00-0000', 'bowler_dob'] = bowler_df['bowler_dob']
```

```python
numerical_cols = bowler_df.select_dtypes(include=['int64', 'float64']).columns
```

```python
numerical_cols
```

```python
# Columns to remove
cols_to_remove = ['match id', 'bowler_id', 'is_batsman_captain', 'is_batsman_keeper', 'inning', 'is_bowler_keeper']

numerical_cols = list(numerical_cols)

# Remove columns from the list
for col in cols_to_remove:
    if col in numerical_cols:
        numerical_cols.remove(col)
```

```python
numerical_cols
```

```python
for col in numerical_cols:
    Q1 = bowler_df[col].quantile(0.25)
    Q3 = bowler_df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = bowler_df[(bowler_df[col] < lower_bound) | (bowler_df[col] > upper_bound)][col]
    print(f"Outliers in column '{col}': {outliers.values}")
```

```python
for col in numerical_cols:
    Q1 = bowler_df[col].quantile(0.25)
    Q3 = bowler_df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    non_outliers = bowler_df[(bowler_df[col] >= lower_bound) & (bowler_df[col] <= upper_bound)][col]
    mean_value = int(np.round(non_outliers.mean()))  # Round off the mean value
    bowler_df.loc[bowler_df[col] < lower_bound, col] = mean_value
    bowler_df.loc[bowler_df[col] > upper_bound, col] = mean_value
```

```python
final_output = 'bowler.xlsx'
bowler_df.to_excel(final_output, index=False)
```
