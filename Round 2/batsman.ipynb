```python
# Replace NaN values with empty strings and ensure all entries are strings
batsman['wicket kind'] = batsman['wicket kind'].fillna('').astype(str)

# Creating 'Out/Not Out' column
batsman['Out_Not_Out'] = batsman['wicket kind'].apply(lambda x: 0 if len(x) == 0 else 1)
```

```python
def calculate_batsman_stats(group):
    total_runs = group['runs'].sum()
    total_balls = group['balls_faced'].sum()
    times_out = group['Out_Not_Out'].sum()
    hundreds = (group['runs'] >= 100).sum()
    fifties = (group['runs'] >= 50).sum()
    return pd.Series({
        'Total Runs': total_runs,
        'Total Balls Faced': total_balls,
        'Total Number of Times Out': times_out,
        "100's": hundreds,
        "50's": fifties
    })
```

```python
# Group by 'batsman_id' and apply the calculation function
batsman_stats = batsman.groupby('batsman_id').apply(calculate_batsman_stats).reset_index()

# Merge the calculated stats back into the original DataFrame
batsman = batsman.merge(batsman_stats, on='batsman_id', how='left')
```

```python
batsman['Strike Rate'] = (batsman['Total Runs'] / batsman['Total Balls Faced']) * 100
batsman['Avg'] = (batsman['Total Runs'] / batsman['Total Number of Times Out'])
```

```python
def strike_rate_weight(sr):
    if sr >= 150:
        return 50
    elif sr >= 100:
        return 40
    elif sr >= 80:
        return 30
    else:
        return 0

def avg_weight(avg):
    if avg >= 50:
        return 30
    elif avg >= 40:
        return 20
    elif avg >= 30:
        return 10
    else:
        return 5

def hundreds_weight(hundreds):
    if hundreds >= 3:
        return 30
    elif hundreds == 2:
        return 20
    elif hundreds == 1:
        return 10
    else:
        return 0

def fifties_weight(fifties):
    if fifties >= 5:
        return 20
    elif fifties >= 3:
        return 10
    elif fifties >= 1:
        return 5
    else:
        return 0

# Apply the functions to the respective columns
batsman['Strike Rate'] = batsman['Strike Rate'].apply(strike_rate_weight)
batsman['Avg'] = batsman['Avg'].apply(avg_weight)
batsman['100s'] = batsman['100s'].apply(hundreds_weight)
batsman['50s'] = batsman['50s'].apply(fifties_weight)
```

```python
batsman.head()
```

```python
batsman['cumulative_points'] = batsman['Strike Rate'] + batsman['Avg'] + batsman['100s'] + batsman['50s']
```

```python
batsman.head()
```

```python
# Step 1: Calculate Consistency Score based on strike rate
batsman['std_strike_rate'] = batsman.groupby('batsman_id')['strike_rate'].transform('std')

# Step 2: Calculate Recency Score using Exponentially Weighted Average for strike rate
batsman = batsman.sort_values(by=['batsman_id', 'match_dt'])
batsman['exp_weighted_avg_sr'] = batsman.groupby('batsman_id')['strike_rate'].transform(lambda x: x.ewm(span=5, adjust=False).mean())

# Normalize consistency and recency scores
batsman['std_strike_rate_norm'] = (batsman['std_strike_rate'] - batsman['std_strike_rate'].min()) / (batsman['std_strike_rate'].max() - batsman['std_strike_rate'].min())
batsman['exp_weighted_avg_sr_norm'] = (batsman['exp_weighted_avg_sr'] - batsman['exp_weighted_avg_sr'].min()) / (batsman['exp_weighted_avg_sr'].max() - batsman['exp_weighted_avg_sr'].min())

# Combine Scores
# Assuming we give equal weight to each component
batsman['final_score'] = (batsman['cumulative_points'] / batsman['cumulative_points'].max()) * 0.7 + \
                         (1 - batsman['std_strike_rate_norm']) * 0.2 + \
                         batsman['exp_weighted_avg_sr_norm'] * 0.1

# Get the final score for each batsman
final_batsman_scores = batsman.groupby('batsman_id')['final_score'].mean().reset_index()

print(final_batsman_scores)
```

```python
import numpy as np
import pandas as pd

def calculate_consistency(group):
    runs = group['runs']
    total_runs = group['Total Runs'].iloc[0]
    times_out = group['Total Number of Times Out'].iloc[0]
    
    # Calculate average per innings
    avg_per_innings = total_runs / len(runs)
    
    # Calculate average per dismissal (this is what's stored in 'Avg')
    avg_per_dismissal = total_runs / times_out if times_out > 0 else total_runs
    
    # Calculate standard deviation of runs
    std_dev = np.std(runs)
    
    # Calculate Coefficient of Variation (CV) using avg_per_innings
    cv = std_dev / avg_per_innings if avg_per_innings > 0 else 0
    
    # Convert CV to a consistency score
    consistency_score = 1 - cv
    
    # Normalize to ensure the score is between 0 and 1
    return pd.Series({'consistency_score': max(0, min(1, consistency_score))})

# Apply the consistency function to each group (each batsman)
batsman['consistency_score'] = batsman.groupby('batsman_id').apply(calculate_consistency).reset_index(level=0, drop=True)

def calculate_recency(group):
    runs = group['runs']
    dates = pd.to_datetime(group['match_dt'])
    half_life = 30  # Number of days after which the weight becomes half
    
    # Calculate time difference from the most recent match for this batsman
    time_diff = (dates.max() - dates).dt.days
    
    # Calculate weights based on time difference
    weights = 0.5 ** (time_diff / half_life)
    
    # Calculate weighted average of runs
    recency_score = np.average(runs, weights=weights)
    
    return recency_score

# Apply the recency function to each group (each batsman)
batsman['recency_score'] = batsman.groupby('batsman_id').apply(calculate_recency)

# Normalize recency score globally after calculating for each batsman
batsman['recency_score'] = (batsman['recency_score'] - batsman['recency_score'].min()) / (batsman['recency_score'].max() - batsman['recency_score'].min())

def calculate_combined_score(row, weights=(0.3, 0.3, 0.4)):
    return weights[0] * row['consistency_score'] + weights[1] * row['recency_score'] + weights[2] * row['cumulative_points']

batsman['combined_score'] = batsman.apply(calculate_combined_score, axis=1)

# Sort batsmen by combined score
ranked_batsmen = batsman.sort_values('combined_score', ascending=False)

# Remove duplicates, keeping the first occurrence of each batsman_id
unique_batsmen = ranked_batsmen.drop_duplicates(subset='batsman_id', keep='first')

# Reset the index if needed
unique_batsmen = unique_batsmen.reset_index(drop=True)
```

```python
final_output = 'batsman_weighted.xlsx'
unique_batsmen.to_excel(final_output, index=False)
```

```python
def calculate_consistency(batsman, batsman_id):
    # Filter data for the specific batsman
    batsman_data = batsman[batsman['batsman_id'] == batsman_id]
    
    # Calculate the standard deviation of runs
    std_dev = batsman_data['runs'].std()
    
    return std_dev

# Create a new column for consistency
batsman['consistency'] = None

# Calculate and assign consistency for each unique batsman_id
for id in batsman['batsman_id'].unique():
    consistency_value = calculate_consistency(batsman, id)
    batsman.loc[batsman['batsman_id'] == id, 'consistency'] = consistency_value
```
